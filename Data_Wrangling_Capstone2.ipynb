# 1. Problem Definition
The problem is defined in the classification framework, where the predicted variable is “Charge-Off ”. A charge-off is a debt that a creditor has given up trying to collect on after you’ve missed payments for several months. The predicted variable takes value 1 in case of charge-off and 0 otherwise.
This project aims to analyze data for loans through 2007-2018Q4 from Lending Club available on Kaggle.

# 3. Data Preparation and Feature Selection

# 3.1. Preparing the predicted variable

```python
# value counts for the variable 'loan_status'
loan['loan_status'].value_counts(dropna=False)
```

```python
# keeping only the features 'Fully Paid' and 'Charge Off'
loan = loan.loc[loan['loan_status'].isin(['Fully Paid', 'Charged Off'])]

#Proportion of the two variables
loan['loan_status'].value_counts(dropna=False)
loan['loan_status'].value_counts(normalize=True, dropna = False)
```

```python
#Count of the two elements
loan['loan_status'].value_counts(dropna=False)
```

```python
loan['loan_status'].head(15)
```

```python
#Making dummy variables (0,1) for 'Fully Paid' and 'Charge Off'
def Default(x):
    if x == 'Fully Paid':
        return 0
    if x == 'Charged Off':
        return 1
```

```python
loan['Default'] = loan['loan_status'].apply(Default)
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
loan.head(20)
```

```python
loan = loan.drop(['loan_status'], axis = 1)
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
loan.head(20)
```

# 3.2. Feature Selection-Limit the Feature Space

# 3.2.1. Features elimination by significant missing values

```python
#Normalized proportions of missing values in each column

missing_fractions = loan.isnull().mean().sort_values(ascending=False)

missing_fractions.head(50)
```

```python
#dropping features that have more than 35% missing values

drop_list = sorted(list(missing_fractions[missing_fractions > 0.35].index))

print(drop_list)
```

```python
len(drop_list)
```

```python
loan.drop(labels=drop_list, axis=1, inplace=True)
loan.shape
```

# 3.2.2. Feature elimination based on intuitiveness

```python
print(sorted(loan.columns))
```

```python
keep_list = ['Default','funded_amnt','addr_state', 'annual_inc', 'application_type', 'dti', 'earliest_cr_line', 'emp_length', 'emp_title', 'fico_range_high', 'fico_range_low', 'grade', 'home_ownership', 'id', 'initial_list_status', 'installment', 'int_rate', 'loan_amnt', 'loan_status', 'mort_acc', 'open_acc', 'pub_rec', 'pub_rec_bankruptcies', 'purpose', 'revol_bal', 'revol_util', 'sub_grade', 'term', 'title', 'total_acc', 'verification_status', 'zip_code','last_pymnt_amnt','num_actv_rev_tl', 'mo_sin_rcnt_rev_tl_op','mo_sin_old_rev_tl_op',"bc_util","bc_open_to_buy","avg_cur_bal","acc_open_past_24mths" ]

len(keep_list)
```

```python
print(loan['Default'].head(20))
```

```python
#Dropping everything except elements in keep_list

drop_list = [col for col in loan.columns if col not in keep_list]

loan.drop(labels=drop_list, axis=1, inplace=True)

loan.shape
```

# 3.2.3. Features elimination based on the correlation

```python
correlation = loan.corr()
correlation_Default = abs(correlation['Default'])
```

```python
correlation_Default.sort_values(ascending=False)
```

```python
#Dropping columns which are less than 3% correlated to 'Default'

drop_list_corr = sorted(list(correlation_Default[correlation_Default < 0.03].index))
print(drop_list_corr)
```

```python
loan.drop(labels=drop_list_corr, axis=1, inplace=True)
loan.shape
```
